```python
import requests

BASE = "https://pds-imaging.jpl.nasa.gov/solr/pds_archives/search"

params = {
    "target": "mars",
    "mission": "mars*science*laboratory",
    "instrument": "mastcam",
    "filter": "red",
}

r = requests.get(BASE, params=params, timeout=30)
r.raise_for_status()
data = r.json()

# Inspect returned docs
docs = data.get("response", {}).get("docs", [])
print("hits:", data.get("response", {}).get("numFound", 0))
print("first keys:", list(docs[0].keys()) if docs else None)
```
